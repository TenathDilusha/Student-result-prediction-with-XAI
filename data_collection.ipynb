```python
import pandas as pd
```

```python
df1 = pd.read_csv('./dataset/Anxiety.csv')
df2 = pd.read_csv('./dataset/Depression.csv')
df3 = pd.read_csv('./dataset/Stress.csv')

pd.set_option('display.max_columns', None)
```

```python
df1 = df1.rename(columns={
    '1. Age': 'Age',
    '2. Gender': 'Gender',
    '3. University': 'University',
    '4. Department': 'Department',
    '5. Academic Year': 'Academic Year',
    '6. Current CGPA': 'Current CGPA',
    '7. Did you receive a waiver or scholarship at your university?': 'Scholarship',
    '1. In a semester, how often you felt nervous, anxious or on edge due to academic pressure? ': 'Nervous',
    '2. In a semester, how often have you been unable to stop worrying about your academic affairs? ': 'Worry',
    '3. In a semester, how often have you had trouble relaxing due to academic pressure? ': 'Trouble_Relaxing',
    '4. In a semester, how often have you been easily annoyed or irritated because of academic pressure?': 'Irritated',
    '5. In a semester, how often have you worried too much about academic affairs? ': 'Overthinking',
    '6. In a semester, how often have you been so restless due to academic pressure that it is hard to sit still?': 'Restless',
    '7. In a semester, how often have you felt afraid, as if something awful might happen?': 'Fear',
    'Anxiety Value': 'Anxiety_Value',
    'Anxiety Label': 'Anxiety_Label'
})
```

```python
df2 = df2.rename(columns={
    '1. Age': 'Age',
    '2. Gender': 'Gender',
    '3. University': 'University',
    '4. Department': 'Department',
    '5. Academic Year': 'Academic Year',
    '6. Current CGPA': 'Current CGPA',
    '7. Did you receive a waiver or scholarship at your university?': 'Scholarship',
    '1. In a semester, how often have you had little interest or pleasure in doing things?': 'Interest_Loss',
    '2. In a semester, how often have you been feeling down, depressed or hopeless?': 'Feeling_Depressed',
    '3. In a semester, how often have you had trouble falling or staying asleep, or sleeping too much? ': 'Sleep_Issues',
    '4. In a semester, how often have you been feeling tired or having little energy? ': 'Fatigue',
    '5. In a semester, how often have you had poor appetite or overeating? ': 'Appetite_Changes',
    '6. In a semester, how often have you been feeling bad about yourself - or that you are a failure or have let yourself or your family down? ': 'Low_Self_Worth',
    '7. In a semester, how often have you been having trouble concentrating on things, such as reading the books or watching television? ': 'Concentration_Issues',
    "8. In a semester, how often have you moved or spoke too slowly for other people to notice? Or you've been moving a lot more than usual because you've been restless? ": 'Psychomotor_Changes',
    '9. In a semester, how often have you had thoughts that you would be better off dead, or of hurting yourself? ': 'Suicidal_Thoughts',
    'Depression Value': 'Depression_Value',
    'Depression Label': 'Depression_Label'
})
```

```python
df3 = df3.rename(columns={
    '1. Age': 'Age',
    '2. Gender': 'Gender',
    '3. University': 'University',
    '4. Department': 'Department',
    '5. Academic Year': 'Academic Year',
    '6. Current CGPA': 'Current CGPA',
    '7. Did you receive a waiver or scholarship at your university?': 'Scholarship',
    '1. In a semester, how often have you felt upset due to something that happened in your academic affairs? ': 'Upset',
    '2. In a semester, how often you felt as if you were unable to control important things in your academic affairs?': 'Lack_Control',
    '3. In a semester, how often you felt nervous and stressed because of academic pressure? ': 'Nervous_Stress',
    '4. In a semester, how often you felt as if you could not cope with all the mandatory academic activities? (e.g, assignments, quiz, exams) ': 'Not_Coping',
    '5. In a semester, how often you felt confident about your ability to handle your academic / university problems?': 'Confident',
    '6. In a semester, how often you felt as if things in your academic life is going on your way? ': 'Life_On_Track',
    '7. In a semester, how often are you able to control irritations in your academic / university affairs? ': 'Control_Irritation',
    '8. In a semester, how often you felt as if your academic performance was on top?': 'Performance',
    '9. In a semester, how often you got angered due to bad performance or low grades that is beyond your control? ': 'Anger',
    '10. In a semester, how often you felt as if academic difficulties are piling up so high that you could not overcome them? ': 'Overwhelmed',
    'Stress Value': 'Stress_Value',
    'Stress Label': 'Stress_Label'
})
```

```python
common_cols = [
    'Age',
    'Gender',
    'University',
    'Department',
    'Academic Year',
    'Current CGPA',
    'Scholarship'
]

df_merged = df1.merge(df2, on=common_cols, how='inner')
df_merged = df_merged.merge(df3, on=common_cols, how='inner')
```
