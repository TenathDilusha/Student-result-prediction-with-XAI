```python
import pandas as pd
```

```python
df1 = pd.read_csv('./dataset/Anxiety.csv')
df2 = pd.read_csv('./dataset/Depression.csv')
df3 = pd.read_csv('./dataset/Stress.csv')
```

```python
def merge_by_row(anx: pd.DataFrame, dep: pd.DataFrame, str_df: pd.DataFrame) -> pd.DataFrame:
    """Align rows by position and combine shared base features plus disorder-specific items."""
    # Trim to common length to keep row alignment
    min_len = min(len(anx), len(dep), len(str_df))
    anx = anx.iloc[:min_len].reset_index(drop=True)
    dep = dep.iloc[:min_len].reset_index(drop=True)
    str_df = str_df.iloc[:min_len].reset_index(drop=True)

    # Shared base (first 7 cols in source) normalized
    base = anx.iloc[:, :7].copy()
    base.columns = [
        "age",
        "gender",
        "university",
        "department",
        "academic_year",
        "cgpa",
        "scholarship",
    ]

    def only_questions(df: pd.DataFrame, prefix: str) -> pd.DataFrame:
        cols = df.columns.tolist()
        q_cols = cols[7:-2]
        mapping = {c: f"{prefix}_q{i+1}" for i, c in enumerate(q_cols)}
        mapping[cols[-2]] = f"{prefix}_score"
        mapping[cols[-1]] = f"{prefix}_label"
        out = df.rename(columns=mapping)
        return out[list(mapping.values())]

    merged = pd.concat(
        [
            base,
            only_questions(anx, "anxiety"),
            only_questions(dep, "depression"),
            only_questions(str_df, "stress"),
        ],
        axis=1,
    )
    return merged

# Build row-aligned combined dataset
row_aligned = merge_by_row(df1, df2, df3)
row_aligned.head()
```
