```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

pd.set_option('display.max_columns', None)
```

## Load merged dataset
Read the combined survey file and show the first few rows to verify columns and types.

```python
df = pd.read_csv('../dataset/merged_dataset.csv')
df.head()
```

## Inspect gender distribution
Check how gender values are represented before cleaning.

## Clean and encode gender
Remove responses marked 'Prefer not to say', one-hot encode remaining gender categories, and preview the updated frame.

```python
df = df[df['Gender'] != 'Prefer not to say'].copy()
df['Gender_Male'] = df['Gender'] == 'Male'
df = df.drop(["Gender"], axis=1)
```

## Encode stress labels
Map the ordinal stress categories to integers for model-friendly features.

```python
mapping = {
    'Low Stress': 0,
    'Moderate Stress': 1,
    'High Perceived Stress': 2
}

df['Stress_Label_Encoded'] = df['Stress_Label'].map(mapping)
```

```python
le_university = LabelEncoder()
df['University_Encoded'] = le_university.fit_transform(df['University'])
df = df.drop(columns=['University'])

le_department = LabelEncoder()
df['Department_Encoded'] = le_department.fit_transform(df['Department'])
df = df.drop(columns=['Department'])
```

```python
mapping = {
    '18-22': 20, 
    '23-26': 24, 
    '27-30': 28, 
    'Above 30': 32, 
    'Below 18': 16}

df['Age_Encoded'] = df['Age'].map(mapping)
df = df.drop(['Age'], axis=1)
```

```python
year_order = {
    'First Year or Equivalent': 1,
    'Second Year or Equivalent': 2,
    'Third Year or Equivalent': 3,
    'Fourth Year or Equivalent': 4,
    'Other': 0  
}

df['Academic_Year_Encoded'] = df['Academic Year'].map(year_order)
df = df.drop(['Academic Year'], axis=1)
```

```python
df = df[df['Current CGPA'] != 'Other']

cgpa_map = {
    'Below 2.50': 2.25,
    '2.50 - 2.99': 2.75,
    '3.00 - 3.39': 3.2,
    '3.40 - 3.79': 3.6,
    '3.80 - 4.00': 3.9
}

df['CGPA_encoded'] = df['Current CGPA'].map(cgpa_map)
```

```python
mapping = {
    'Low Stress': 0,
    'Moderate Stress': 1,
    'High Perceived Stress': 2
}

df['Stress_Label_Encoded'] = df['Stress_Label'].map(mapping)
df = df.drop(columns=['Anxiety_Label', 'Depression_Label', 'Stress_Value', 'Current CGPA', 'Stress_Label'])
```

```python
df['Scholarship_Encoded'] = df['Scholarship'] == 'Yes'
df = df.drop(["Scholarship"], axis=1)
```
