```python
import numpy as np

# Basic dataset checks
target = 'defect'
feature_cols = [c for c in df.columns if c != target]

rows, cols = df.shape
print(f"Rows: {rows:,}, Columns: {cols}")
print(f"Feature columns: {len(feature_cols)}")

missing = (df.isna().sum().sort_values(ascending=False))
print("\nMissing values (top 10):")
print(missing.head(10))

print("\nTarget distribution (counts):")
print(df[target].value_counts())
print("\nTarget distribution (proportions):")
print(df[target].value_counts(normalize=True))

# Correlation with target (top 10 by absolute value)
if np.issubdtype(df[target].dtype, np.number):
    corr = (
        df[feature_cols + [target]].corr()[target]
        .drop(target)
        .sort_values(key=np.abs, ascending=False)
    )
    print("\nTop correlations with target:")
    print(corr.head(10))
else:
    print("Target is non-numeric; skipping correlation." )
```

## Modeling Setup: Train/Test Split and Cross-Validation

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report

random_state = 42
```

```python
X = df.drop(columns=[target])
y = df[target]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=random_state
)

print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")
print(f"Positive rate (train): {y_train.mean():.3f}, (test): {y_test.mean():.3f}")
```

```python
log_reg = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(max_iter=200, n_jobs=1, solver="lbfgs", random_state=random_state)),
])

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)
cv_scores = cross_val_score(
    log_reg, X, y, cv=cv, scoring="roc_auc", n_jobs=-1
)

print(f"LogReg CV ROC-AUC: {cv_scores.mean():.3f} ± {cv_scores.std():.3f}")
```

```python
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)
y_proba_lr = log_reg.predict_proba(X_test)[:, 1]

def summarize_performance(y_true, y_pred, y_proba):
    metrics = {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred),
        "recall": recall_score(y_true, y_pred),
        "f1": f1_score(y_true, y_pred),
        "roc_auc": roc_auc_score(y_true, y_proba),
    }
    for k, v in metrics.items():
        print(f"{k}: {v:.3f}")

print("Logistic Regression (test set):")
summarize_performance(y_test, y_pred_lr, y_proba_lr)

print("\nClassification report:")
print(classification_report(y_test, y_pred_lr))
```

```python
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    n_jobs=1,
    random_state=random_state,
    class_weight="balanced_subsample",
)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
y_proba_rf = rf.predict_proba(X_test)[:, 1]

print("Random Forest (test set):")
summarize_performance(y_test, y_pred_rf, y_proba_rf)
```

```python
rf_importances = (
    pd.Series(rf.feature_importances_, index=X_train.columns)
    .sort_values(ascending=False)
)

print("Top 10 features by importance (Random Forest):")
display(rf_importances.head(10))
```

## Overfitting Check and Model Optimization

```python
# Overfitting check: compare train vs test metrics
print("LogReg Train:")
y_pred_lr_train = log_reg.predict(X_train)
y_proba_lr_train = log_reg.predict_proba(X_train)[:, 1]
summarize_performance(y_train, y_pred_lr_train, y_proba_lr_train)

print("\nLogReg Test:")
summarize_performance(y_test, y_pred_lr, y_proba_lr)

print("\nRandom Forest Train:")
y_pred_rf_train = rf.predict(X_train)
y_proba_rf_train = rf.predict_proba(X_train)[:, 1]
summarize_performance(y_train, y_pred_rf_train, y_proba_rf_train)

print("\nRandom Forest Test:")
summarize_performance(y_test, y_pred_rf, y_proba_rf)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Cross-validated ROC-AUC for baseline RF to gauge generalization
cv_scores_rf = cross_val_score(rf, X, y, cv=cv, scoring="roc_auc", n_jobs=-1)
print(f"Random Forest CV ROC-AUC: {cv_scores_rf.mean():.3f} ± {cv_scores_rf.std():.3f}")
```

```python
# Hyperparameter tuning: Random Forest (light RandomizedSearchCV to avoid heavy compute)
rf_base = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    n_jobs=1,
    random_state=random_state,
    class_weight="balanced_subsample",
)
param_dist = {
    "n_estimators": [120, 200, 280],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2"],
}
rf_search = RandomizedSearchCV(
    rf_base,
    param_distributions=param_dist,
    n_iter=12,
    cv=3,
    scoring="roc_auc",
    n_jobs=-1,
    random_state=random_state,
    verbose=1,
)
rf_search.fit(X_train, y_train)

print("Best RF params:")
print(rf_search.best_params_)
print(f"Best CV ROC-AUC: {rf_search.best_score_:.3f}")

rf_best = rf_search.best_estimator_
y_pred_rf_best = rf_best.predict(X_test)
y_proba_rf_best = rf_best.predict_proba(X_test)[:, 1]

print("\nTuned Random Forest (test set):")
summarize_performance(y_test, y_pred_rf_best, y_proba_rf_best)
```

```python
# Feature importance from tuned RF
rf_best_importances = (
    pd.Series(rf_best.feature_importances_, index=X_train.columns)
    .sort_values(ascending=False)
)
print("Top 10 features by importance (Tuned RF):")
display(rf_best_importances.head(10))
```
